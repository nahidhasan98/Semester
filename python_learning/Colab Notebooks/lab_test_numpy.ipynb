```
import numpy as np

def arrays(arr):
  return np.array(arr[::-1],float)

arr = input().strip().split(' ')
result = arrays(arr)
print(result)
```

```
import numpy as np

np.set_printoptions(sign=' ')

arr = np.array(input().split(),float)

print(np.floor(arr))
print(np.ceil(arr))
print(np.rint(arr))
```

```
import numpy as np

n,m = input().split()

arr = []
for i in range(int(n)):
  arr.append(np.array(input().split(),int))

arr2 = np.sum(arr, axis = 0)

print (np.prod(arr2)) 
```

```
import numpy as np

n,m = input().strip().split()

lst = []

for i in range(int(n)):
  arr = input().strip().split()
  lst.append(arr)

arr2 = np.array(lst,int)

print(arr2.transpose())
print(arr2.flatten())
```
